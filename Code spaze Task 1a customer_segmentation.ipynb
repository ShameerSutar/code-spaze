# Preprocessing

**Encoding Categorical Variables**

```python
from sklearn.preprocessing import LabelEncoder

# Assign the specified mappings for 'gender' and 'preferred_category'
gender_mapping = {'Female': 0, 'Male': 1, 'Other': 2}
category_mapping = {'Clothing': 0, 'Electronics': 1, 'Groceries': 2, 'Home & Garden': 3, 'Sports': 4}

data['gender'] = data['gender'].map(gender_mapping)
data['preferred_category'] = data['preferred_category'].map(category_mapping)
```

```python
data
```

# Clustering
## K-means Clustering

```python
from sklearn.cluster import KMeans

# Elbow Method to find optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Applying K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(data)
data['Cluster'] = clusters
```

**The plot indicates 4 as the optimal number of clusters for segmentation. This suggests that the customer base can be effectively divided into four distinct groups.**

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(data)
data['Cluster'] = clusters
```

**Cluster summary**

```python
# Numerical cluster summary
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)

# Heatmap for cluster summary
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_summary, annot=True, fmt=".2f", cmap='viridis')
plt.title('Cluster Summary')
plt.xlabel('Features')
plt.ylabel('Cluster')
plt.show()
```

**Each cluster shows distinct characteristics:**

> Cluster 0: Moderate income and spending score.
    
> Cluster 1: Low income and moderate spending score.
    
> Cluster 2: High income and high spending score.
    
> Cluster 3: Moderate income and high spending score.
